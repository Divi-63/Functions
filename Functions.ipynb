**Theory Questions:**

1. What is the difference between a function and a method in Python?

- A function is defined independently using def.

- A method is a function that belongs to an object and is called on that object.

2. What is the concept of function arguments and parameters in Python?

- Parameters are variables listed in the function definition.

- Arguments are the actual values passed to the function when it is called.

3. What are the different ways to define and call a function in Python?

- Functions are defined using def or lambda.

- They are called using function_name() with or without arguments.

4. What is the purpose of the return statement in a Python function?

- The return statement ends the function and sends the result back to the caller.

5. What are iterators in Python and how do they differ from iterables?

- An iterator is an object with a __next__() method that returns items one by one.

- An iterable is any object that can be looped over (e.g., list, tuple) and can return an iterator using iter().

6. What is the concept of generators in Python and how are they defined?

- Generators are functions that yield values using the yield keyword.

- They return an iterator and generate values lazily, one at a time.

7. What are the advantages of using generators over regular functions?

- They are memory-efficient, support lazy evaluation, and are ideal for handling large datasets or infinite sequences.

8. What is a lambda function in Python and when is it typically used?

- A lambda is an anonymous, one-line function: lambda x: x + 1.

- Commonly used in short-term operations like with map() or filter().

9. What is the purpose and usage of the map() function in Python?

- map() applies a given function to all items in an iterable: map(func, iterable).

10. What is the difference between map(), reduce(), and filter() functions in Python?

- map(): Applies a function to each item.

- filter(): Keeps items that meet a condition.

- reduce(): Combines all items into a single value (requires functools).

```python
#1. Write a Python function that takes a list of numbers and returns the sum of all even numbers.
def sum_even(numbers):
    return sum(n for n in numbers if n % 2 == 0)

# Example: sum_even([1, 2, 3, 4]) → 6
```

```python
#2. Create a Python function that accepts a string and returns the reverse.
def reverse_string(s):
    return s[::-1]

# Example: reverse_string("hello") → "olleh"
```

```python
#3. Implement a function that takes a list and returns a list of squared values.
def square_list(nums):
    return [n**2 for n in nums]

# Example: square_list([1, 2, 3]) → [1, 4, 9]
```

```python
#4. Write a function to check if a number (1–200) is prime.
def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0: return False
    return True

# Example: is_prime(7) → True


```

```python
#5. Create an iterator class for Fibonacci sequence up to n terms.
class Fibonacci:
    def __init__(self, n):
        self.a, self.b, self.n = 0, 1, n

    def __iter__(self):
        self.count = 0
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return self.a

# Example: list(Fibonacci(5)) → [1, 1, 2, 3, 5]
```

```python
#6. Write a generator that yields powers of 2 up to a given exponent.
def powers_of_two(exp):
    for i in range(exp + 1):
        yield 2 ** i

# Example: list(powers_of_two(3)) → [1, 2, 4, 8]
```

```python
#7. Generator to read a file line by line.
def read_lines(filename):
    with open(filename, 'r') as f:
        for line in f:
            yield line.strip()

# Usage:
# for line in read_lines('file.txt'):
#     print(line)
```

```python
#8. Use a lambda to sort a list of tuples by the second element.
data = [(1, 3), (2, 1), (4, 2)]
sorted_data = sorted(data, key=lambda x: x[1])

# Output: [(2, 1), (4, 2), (1, 3)]
```

```python
#9. Use map() to convert Celsius to Fahrenheit.
celsius = [0, 10, 20]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

# Output: [32.0, 50.0, 68.0]
```
